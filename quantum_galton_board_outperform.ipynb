````python
import numpy as np
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram, circuit_drawer
from qiskit.circuit.library import QFT
from qiskit.circuit import Parameter
import matplotlib.pyplot as plt
from collections import Counter
import warnings

plt.style.use('default')

# Disable Qiskit warnings
warnings.filterwarnings('ignore', category=DeprecationWarning)
warnings.filterwarnings('ignore', category=FutureWarning)
warnings.filterwarnings('ignore', category=UserWarning)
import logging
logging.getLogger('qiskit').setLevel(logging.ERROR)

# MAIN CONFIGURATION - Change these parameters to modify all circuits
n_levels = 13        # Number of levels for basic circuits (H, CNOT, Realistic)
n_qubits = 13        # Number of qubits for advanced circuits (Exponential, Amplitude, Quantum Walk)
shots = 18192        # Number of simulation shots
````

````python
def create_simple_galton_board(n_levels):
    """
    Simple Galton board - only H gates
    Each qubit is independent -> no quantum correlations
    """
    qreg = QuantumRegister(n_levels, 'q')
    creg = ClassicalRegister(n_levels, 'c')
    qc = QuantumCircuit(qreg, creg, name='Simple_Galton')

    for i in range(n_levels):
        qc.h(qreg[i])

    qc.measure(qreg, creg)
    return qc

simple_circuit = create_simple_galton_board(n_levels)
print("SIMPLE VERSION - only H gates:")
simple_circuit.draw()
````

````python
def create_advanced_galton_board(n_levels, add_phase_gates=True):
    """
    Advanced Galton board with CNOT gates and phase rotations
    
    Args:
        n_levels: number of levels
        add_phase_gates: add phase rotations for complex interference
    """
    position_qubits = n_levels
    helper_qubits = n_levels - 1  # Auxiliary qubits for CNOT operations
    
    total_qubits = position_qubits + helper_qubits
    
    qreg = QuantumRegister(total_qubits, 'q')
    creg = ClassicalRegister(position_qubits, 'c')  # Measure only position qubits
    qc = QuantumCircuit(qreg, creg, name='Advanced_Galton')
    
    # Initialization - particle starts in superposition
    qc.h(qreg[0])
    
    # Create cascade of CNOT gates simulating pegs
    for level in range(n_levels - 1):
        # Current position influences next level
        qc.cx(qreg[level], qreg[level + 1])
        
        # Add rotation for complex dynamics
        if add_phase_gates:
            angle = np.pi / (2 ** (level + 2))
            qc.ry(angle, qreg[level + 1])
        
        # Auxiliary qubit for additional complications
        if level < helper_qubits:
            helper_idx = position_qubits + level
            qc.h(qreg[helper_idx])
            qc.cx(qreg[helper_idx], qreg[level + 1])
            
            # Phase rotation dependent on auxiliary qubit
            if add_phase_gates:
                qc.cz(qreg[level], qreg[helper_idx])
    
    # Additional layer for simulating multiple pegs per level
    for level in range(1, n_levels):
        if level < n_levels - 1:
            # Each qubit influences its neighbor
            qc.cx(qreg[level], qreg[level + 1])
        
        if add_phase_gates:
            # Controlled Z rotation for phase correlations  
            phase = np.pi / (4 * level + 1)
            qc.rz(phase, qreg[level])
    
    # Final Hadamard gates for mixing
    for i in range(min(3, n_levels)):
        qc.h(qreg[i])

    for i in range(position_qubits):
        qc.measure(qreg[i], creg[i])
    
    return qc

advanced_circuit = create_advanced_galton_board(n_levels)
print("ADVANCED VERSION with CNOT gates:")
advanced_circuit.draw()
````

````python
def create_realistic_galton_board(n_levels):
    """
    Realistic quantum Galton board
    Simulates real physical process with quantum correlations
    """
    # One qubit for each decision level
    qreg = QuantumRegister(n_levels, 'position')
    # Auxiliary qubits for intermediate calculations
    anc_reg = QuantumRegister(n_levels - 1, 'ancilla') 
    creg = ClassicalRegister(n_levels, 'c')
    
    qc = QuantumCircuit(qreg, anc_reg, creg, name='Realistic_Galton')
    
    # Particle starts in defined state
    qc.x(qreg[0])  # Start with certainty at first position
    
    # Simulate passage through pegs level by level
    for level in range(n_levels - 1):
        # Decision at current level affects next
        qc.h(anc_reg[level])  # Random decision in auxiliary qubit
        
        # If auxiliary qubit is |1⟩, transfer to next level
        qc.ccx(qreg[level], anc_reg[level], qreg[level + 1])
        
        # More complex logic - particle can stay or continue
        qc.cx(anc_reg[level], qreg[level])
        
        # Phase rotations for realistic interference
        angle = np.pi * (level + 1) / (2 * n_levels)
        qc.cry(angle, qreg[level], qreg[level + 1] if level < n_levels - 1 else qreg[0])
        
        # Partial decoupling
        qc.h(qreg[level])
    
    # Final mixing
    for i in range(n_levels):
        qc.ry(np.pi/8, qreg[i])

    qc.measure(qreg, creg)
    
    return qc

realistic_circuit = create_realistic_galton_board(n_levels)
print("REALISTIC VERSION:")
print(realistic_circuit.draw())
````

````python
def simulate_and_analyze(circuit, shots=4096, name="Circuit"):
    """Simulates circuit and analyzes results"""
    
    simulator = AerSimulator()
    
    # Transpilation for optimization
    transpiled = transpile(circuit, simulator, optimization_level=2)
    
    # Simulation
    job = simulator.run(transpiled, shots=shots)
    result = job.result()
    counts = result.get_counts()
    
    # Convert to positions
    positions = {}
    n_qubits = len(list(counts.keys())[0])
    
    for bitstring, count in counts.items():
        # Number of ones = position to the right from center
        ones = bitstring.count('1')
        position = ones - n_qubits//2  # Centering around 0
        positions[position] = positions.get(position, 0) + count
    
    return counts, positions, transpiled.depth()

print(f"Running simulations for {n_levels} levels, {shots} shots...")
print("=" * 60)

# Simulate all variants
simple_counts, simple_pos, simple_depth = simulate_and_analyze(
    create_simple_galton_board(n_levels), shots, "Simple"
)

advanced_counts, advanced_pos, advanced_depth = simulate_and_analyze(
    create_advanced_galton_board(n_levels), shots, "Advanced" 
)

realistic_counts, realistic_pos, realistic_depth = simulate_and_analyze(
    create_realistic_galton_board(n_levels), shots, "Realistic"
)

print(f"Circuit depths:")
print(f"  Simple:    {simple_depth}")
print(f"  Advanced:  {advanced_depth}")  
print(f"  Realistic: {realistic_depth}")
````

````python
def plot_comparison(simple_pos, advanced_pos, realistic_pos):
    """Comparison of all three distributions"""
    
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    
    # Individual histograms
    datasets = [
        (simple_pos, "Simple (H gates only)", "lightblue"),
        (advanced_pos, "Advanced (with CNOT)", "lightcoral"), 
        (realistic_pos, "Realistic", "lightgreen")
    ]
    
    for idx, (data, title, color) in enumerate(datasets):
        ax = axes[idx//2, idx%2]
        positions = sorted(data.keys())
        counts = [data[pos] for pos in positions]
        
        ax.bar(positions, counts, color=color, alpha=0.7, edgecolor='black')
        ax.set_title(title, fontsize=12, fontweight='bold')
        ax.set_xlabel('Position')
        ax.set_ylabel('Particle count')
        ax.grid(True, alpha=0.3)
        
        # Statistics
        total = sum(counts)
        mean_pos = sum(pos * count for pos, count in data.items()) / total
        ax.text(0.02, 0.98, f'Mean: {mean_pos:.2f}', 
                transform=ax.transAxes, verticalalignment='top',
                bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
    
    # Comparison of all three
    ax = axes[1, 1]
    
    all_positions = set()
    all_positions.update(simple_pos.keys())
    all_positions.update(advanced_pos.keys()) 
    all_positions.update(realistic_pos.keys())
    all_positions = sorted(all_positions)
    
    simple_counts = [simple_pos.get(pos, 0) for pos in all_positions]
    advanced_counts = [advanced_pos.get(pos, 0) for pos in all_positions]
    realistic_counts = [realistic_pos.get(pos, 0) for pos in all_positions]
    
    x = np.arange(len(all_positions))
    width = 0.25
    
    ax.bar(x - width, simple_counts, width, label='Simple', alpha=0.7, color=datasets[0][2])
    ax.bar(x, advanced_counts, width, label='Advanced', alpha=0.7, color=datasets[1][2])
    ax.bar(x + width, realistic_counts, width, label='Realistic', alpha=0.7, color=datasets[2][2])
    
    ax.set_xlabel('Position')
    ax.set_ylabel('Particle count')
    ax.set_title('Comparison of All Variants')
    ax.set_xticks(x)
    ax.set_xticklabels(all_positions)
    ax.legend()
    ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

plot_comparison(simple_pos, advanced_pos, realistic_pos)
````

````python
def calculate_statistics(positions, name):
    """Calculates distribution statistics"""
    
    total_particles = sum(positions.values())
    
    # Mean
    mean = sum(pos * count for pos, count in positions.items()) / total_particles
    
    # Variance
    variance = sum((pos - mean)**2 * count for pos, count in positions.items()) / total_particles
    std = np.sqrt(variance)
    
    # Entropy (measure of "randomness")
    entropy = -sum((count/total_particles) * np.log2(count/total_particles) 
                   for count in positions.values() if count > 0)
    
    return {
        'name': name,
        'mean': mean,
        'std': std,
        'variance': variance,
        'entropy': entropy,
        'total_states': len(positions),
        'max_prob': max(positions.values()) / total_particles
    }

stats = [
    calculate_statistics(simple_pos, "Simple"),
    calculate_statistics(advanced_pos, "Advanced"), 
    calculate_statistics(realistic_pos, "Realistic")
]

print("STATISTICAL COMPARISON:")
print("=" * 80)
print(f"{'Version':<12} {'Mean':<8} {'Std Dev':<10} {'Variance':<8} {'Entropy':<8} {'States':<6} {'Max prob':<8}")
print("-" * 80)

for s in stats:
    print(f"{s['name']:<12} {s['mean']:<8.2f} {s['std']:<10.2f} {s['variance']:<8.2f} "
          f"{s['entropy']:<8.2f} {s['total_states']:<6} {s['max_prob']:<8.3f}")
````

````python
def analyze_quantum_properties(circuit, name):
    """Analyzes quantum properties of the circuit"""
    
    # Number of quantum gates
    gate_counts = circuit.count_ops()
    
    # Circuit depth
    depth = circuit.depth()
    
    # Number of CNOT gates (measure of entanglement)
    cnot_count = gate_counts.get('cx', 0) + gate_counts.get('ccx', 0)
    
    # Total number of two-qubit gates
    two_qubit_gates = cnot_count + gate_counts.get('cz', 0) + gate_counts.get('cy', 0)
    
    print(f"\n{name}:")
    print(f"  Circuit depth: {depth}")
    print(f"  CNOT gates: {cnot_count}")
    print(f"  Total 2-qubit gates: {two_qubit_gates}")
    print(f"  Hadamard gates: {gate_counts.get('h', 0)}")
    print(f"  Rotation gates: {gate_counts.get('ry', 0) + gate_counts.get('rz', 0)}")

print("QUANTUM PROPERTIES ANALYSIS OF CIRCUITS:")
print("=" * 50)

analyze_quantum_properties(create_simple_galton_board(n_levels), "Simple version")
analyze_quantum_properties(create_advanced_galton_board(n_levels), "Advanced version") 
analyze_quantum_properties(create_realistic_galton_board(n_levels), "Realistic version")
````

````python
def create_exponential_galton_board(n_qubits, sigma=1.0):
    """
    Exponential approximation of quantum Galton board
    Uses QFT and phase rotations for Gaussian distribution approximation
    
    Args:
        n_qubits: number of qubits (determines resolution)
        sigma: standard deviation of Gaussian distribution
    """
    qreg = QuantumRegister(n_qubits, 'q')
    creg = ClassicalRegister(n_qubits, 'c')
    qc = QuantumCircuit(qreg, creg, name='Exponential_Galton')
    
    # Create uniform superposition of all states
    for i in range(n_qubits):
        qc.h(qreg[i])
    
    # Apply QFT for transition to frequency domain
    qft = QFT(n_qubits, approximation_degree=0, insert_barriers=False)
    qc.append(qft, qreg)
    
    # Phase rotations for Gaussian distribution approximation
    # Exponential approximation: exp(-x²/(2σ²))
    for i in range(n_qubits):
        # Each qubit represents different frequency
        frequency = i / (2**n_qubits)
        
        # Gaussian phase: -π * frequency² / (2 * sigma²)
        phase = -np.pi * (frequency**2) / (2 * sigma**2)
        qc.p(phase, qreg[i])
        
        # Additional phase correlations between qubits
        for j in range(i+1, min(i+3, n_qubits)):  # Limited correlations
            cross_phase = -np.pi * frequency * (j/(2**n_qubits)) / (4 * sigma**2)
            qc.cp(cross_phase, qreg[i], qreg[j])
    
    # Inverse QFT for return to position domain
    qft_inv = QFT(n_qubits, approximation_degree=0, insert_barriers=False, inverse=True)
    qc.append(qft_inv, qreg)

    qc.measure(qreg, creg)
    
    return qc
````

````python
def create_variational_galton_board(n_qubits, n_layers=3):
    """
    Variational quantum Galton board with parameterized gates
    Allows optimization of parameters for specific target distribution
    """
    qreg = QuantumRegister(n_qubits, 'q')
    creg = ClassicalRegister(n_qubits, 'c')
    qc = QuantumCircuit(qreg, creg, name='Variational_Galton')
    
    # Parameters for optimization
    params = []
    
    # Initialization
    for i in range(n_qubits):
        qc.h(qreg[i])
    
    # Variational layers
    for layer in range(n_layers):
        # Rotation gates with parameters
        for i in range(n_qubits):
            theta = Parameter(f'θ_{layer}_{i}')
            phi = Parameter(f'φ_{layer}_{i}')
            params.extend([theta, phi])
            
            qc.ry(theta, qreg[i])
            qc.rz(phi, qreg[i])
        
        # Entanglement with neighbors
        for i in range(n_qubits - 1):
            alpha = Parameter(f'α_{layer}_{i}')
            params.append(alpha)
            
            qc.cx(qreg[i], qreg[i+1])
            qc.rz(alpha, qreg[i+1])
            qc.cx(qreg[i], qreg[i+1])
        
        # Global phase gate
        if layer < n_layers - 1:
            for i in range(0, n_qubits-1, 2):
                qc.cz(qreg[i], qreg[i+1])
    
    # Final adjustment
    for i in range(n_qubits):
        final_angle = Parameter(f'final_{i}')
        params.append(final_angle)
        qc.ry(final_angle, qreg[i])
    
    qc.measure(qreg, creg)
    
    return qc, params
````

````python
def create_amplitude_encoding_galton(n_qubits, target_distribution=None):
    """
    Amplitude encoding approach - directly encodes target distribution into amplitudes
    """
    if target_distribution is None:
        # Default: Gaussian distribution
        x = np.linspace(-3, 3, 2**n_qubits)
        target_distribution = np.exp(-x**2/2) / np.sqrt(2*np.pi)
        target_distribution = target_distribution / np.linalg.norm(target_distribution)
    
    qreg = QuantumRegister(n_qubits, 'q')
    creg = ClassicalRegister(n_qubits, 'c')
    qc = QuantumCircuit(qreg, creg, name='Amplitude_Encoded_Galton')
    
    # Initialize all qubits to |0⟩
    qc.initialize(target_distribution, qreg)
    
    # Add some quantum dynamics
    #for i in range(n_qubits//2):
    #    qc.ry(np.pi/8, qreg[i])
    #    if i < n_qubits - 1:
    #        qc.cx(qreg[i], qreg[i+1])

    qc.measure(qreg, creg)
    
    return qc
````

````python
def create_quantum_walk_galton(n_qubits, steps=None):
    """
    Quantum Random Walk implementation of Galton board
    Uses quantum random walk to simulate particle movement
    """
    if steps is None:
        steps = n_qubits * 2
    
    # Position qubits + coin qubit
    pos_qubits = n_qubits
    coin_qubit = 1
    total_qubits = pos_qubits + coin_qubit
    
    qreg = QuantumRegister(total_qubits, 'q')
    creg = ClassicalRegister(pos_qubits, 'c')
    qc = QuantumCircuit(qreg, creg, name='Quantum_Walk_Galton')
    
    # Initialization - particle in center, coin in superposition
    center = pos_qubits // 2
    if center < pos_qubits:
        qc.x(qreg[center])  # Position in center
    qc.h(qreg[pos_qubits])  # Coin qubit in superposition
    
    # Quantum walk steps
    for step in range(steps):
        # Coin flip
        angle = np.pi/3 + step * np.pi/20  # Changing bias
        qc.ry(angle, qreg[pos_qubits])
        
        # Conditional shifts
        for pos in range(pos_qubits - 1):
            # If coin = |0⟩, shift left
            qc.x(qreg[pos_qubits])  # Flip coin
            qc.ccx(qreg[pos_qubits], qreg[pos+1], qreg[pos])
            qc.x(qreg[pos_qubits])  # Flip back
            
            # If coin = |1⟩, shift right  
            qc.ccx(qreg[pos_qubits], qreg[pos], qreg[pos+1])
        
        # Entanglement between position and coin
        for pos in range(0, pos_qubits, 2):
            qc.cz(qreg[pos], qreg[pos_qubits])
    
    # Final decomposition
    qc.h(qreg[pos_qubits])

    qc.measure(qreg[:pos_qubits], creg)
    
    return qc
````

````python
def simulate_and_analyze_amplitude(circuit, shots=4096, name="Circuit"):
    simulator = AerSimulator()
    
    # Transpilation for optimization
    transpiled = transpile(circuit, simulator, optimization_level=2)
    
    # Simulation
    job = simulator.run(transpiled, shots=shots)
    result = job.result()
    counts = result.get_counts()
    
    # Convert to positions
    positions = {}
    n_qubits = len(list(counts.keys())[0])
    
    for bitstring, count in counts.items():
        # Number of ones = position to the right from center
        ones = bitstring.count('1')
        position = ones - n_qubits//2 - 1.0 
        positions[position] = positions.get(position, 0) + count
    
    return counts, positions, transpiled.depth()
````

````python
# 1. **Quantum Fourier Transform (QFT)**:
#    - Converts between position and momentum representation
#    - Allows working with frequency components of distribution
# 
# 2. **Exponential approximation of Gaussian distribution**:
#    - Uses phase rotations P(φ) = e^(iφ) 
#    - Approximates G(x) ≈ exp(-x²/(2σ²)) using quantum phases
# 
# 3. **Amplitude Encoding**:
#    - Directly encodes target distribution into quantum amplitudes
#    - |ψ⟩ = Σᵢ √(pᵢ) |i⟩, where pᵢ is probability of state i
# 
# 4. **Quantum Random Walk**:
#    - Quantum version of random walk
#    - Combines "coin flip" qubit with position qubits
#    - Exhibits quantum interference => non-classical distributions

advanced_stats = [
    calculate_statistics(exp_pos, "Exponential"),
    calculate_statistics(amp_pos, "Amplitude"), 
    calculate_statistics(walk_pos, "Quantum Walk")
]

print("\nSTATISTICS OF ADVANCED IMPLEMENTATIONS:")
print(f"{'Version':<15} {'Mean':<8} {'Std Dev':<10} {'Variance':<8} {'Entropy':<8} {'States':<6} {'Max prob':<8}")


for s in advanced_stats:
    print(f"{s['name']:<15} {s['mean']:<8.2f} {s['std']:<10.2f} {s['variance']:<8.2f} "
          f"{s['entropy']:<8.2f} {s['total_states']:<6} {s['max_prob']:<8.3f}")

def compare_to_classical_gaussian(positions, n_qubits, name):
    """Compares quantum distribution with classical Gaussian distribution"""
    
    # Classical Gaussian distribution
    x_range = list(range(-n_qubits//2, n_qubits//2 + 1))
    classical_gaussian = []
    sigma = 1.5
    
    for x in x_range:
        prob = np.exp(-x**2 / (2*sigma**2)) / np.sqrt(2*np.pi*sigma**2)
        classical_gaussian.append(prob)
    
    # Normalization
    classical_gaussian = np.array(classical_gaussian)
    classical_gaussian = classical_gaussian / np.sum(classical_gaussian)
    
    # Quantum distribution
    total_shots = sum(positions.values())
    quantum_probs = []
    
    for x in x_range:
        count = positions.get(x, 0)
        quantum_probs.append(count / total_shots)
    
    # Kolmogorov-Smirnov distance
    ks_distance = np.max(np.abs(np.cumsum(classical_gaussian) - np.cumsum(quantum_probs)))
    
    # Chi-square test
    expected = classical_gaussian * total_shots
    observed = [positions.get(x, 0) for x in x_range]
    chi_square = np.sum((np.array(observed) - expected)**2 / (expected + 1e-10))
    
    print(f"\n{name} vs Classical Gaussian:")
    print(f"  KS distance: {ks_distance:.4f}")
    print(f"  Chi-square:  {chi_square:.4f}")
    
    return ks_distance, chi_square

print("")
print("COMPARISON WITH CLASSICAL GAUSSIAN DISTRIBUTION:")

compare_to_classical_gaussian(exp_pos, n_qubits, "Exponential approximation")
compare_to_classical_gaussian(amp_pos, n_qubits, "Amplitude encoding")
compare_to_classical_gaussian(walk_pos, n_qubits, "Quantum walk")
````
